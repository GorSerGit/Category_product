С помощью скачанных выше библиотек приводим значения параметра INFO к основам слов,  и убирем все служебные символы

Пример работы функции показан ниже.
Вход - привет, меня зовут Сергей я бегу 777
Выход - привет я сергей я бег 777

```python
import re
import nltk
from nltk.tokenize import word_tokenize
import pymorphy2

def clean_text(text):
    # Удаление служебных символов
    cleaned_text = re.sub(r'[^а-яА-Я0-9 ]', '', text)

    # Токенизация текста на слова
    words = word_tokenize(cleaned_text)

    morph = pymorphy2.MorphAnalyzer()

    cleaned_words = []
    for word in words:
        # Получение основы слова
        base_form = morph.parse(word)[0].normal_form

        # Исключение предлогов и союзов
        if morph.parse(base_form)[0].tag.POS not in ['PREP', 'CONJ']:
            cleaned_words.append(base_form)

    cleaned_words = set(cleaned_words)

    cleaned_text = ' '.join(cleaned_words)

    return cleaned_text

# Пример использования
text = "привет меня зовут Сергей я бегу 777"
cleaned_text = clean_text(text)
print(cleaned_text)
```

##  **Модель номер 1**



SVM модель. Возвращает только один ответ - тот в котором наиболее уверена.
Точность 0.94

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Разделяем данные на обучающий и тестовый набор
X_train, X_test, y_train, y_test = train_test_split(Train['info'], Train['cat'], test_size=0.01, random_state=42)

# Преобразовываем текст в числовой формат с помощью TF-IDF векторизации
vectorizer = TfidfVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Обучаем модель SVM
svc = SVC(kernel='linear')
svc.fit(X_train_vect, y_train)
```

```python
# Прогнозируем классы для тестового набора
y_pred = svc.predict(X_test_vect)

# Оцениваем точность модели
accuracy = accuracy_score(y_test, y_pred)
print("Точность модели: {:.2f}".format(accuracy))
```

```python
import joblib

# Сохраняем обученную модель
joblib.dump(svc, 'svm_model_2_1.pkl')
```
